```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/kathanbhavsar/Desktop/Python,DS,ML/DATA/cancer_tumor_data_features.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(df)
```

```python
from sklearn.decomposition import PCA
```

```python
pca_model = PCA(n_components=2)
```

```python
pc = pca_model.fit_transform(scaled_X)
```

```python
plt.figure(dpi=200)
plt.scatter(pc[:,0],pc[:,1],c=)
```

```python
pca_model.components_
```

```python
df_comp = pd.DataFrame(pca_model.components_,index = ['PC1','PC2'],columns=df.columns)
```

```python
plt.figure(figsize=(20,3),dpi=200)
sns.heatmap(df_comp)
```

```python
pca_model.explained_variance_ratio_
```

```python
np.sum(pca_model.explained_variance_ratio_)
```

```python
explained_variance = []
for n in  range(1,30):
    pca = PCA(n_components=n)
    pca.fit(scaled_X)
    explained_variance.append(np.sum(pca.explained_variance_ratio_))
```

```python
plt.plot(range(1,30),explained_variance)
plt.ylabel('Variance Explained')
plt.xlabel('Number of components')
```
