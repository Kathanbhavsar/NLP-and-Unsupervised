```python
X = pd.get_dummies(df)
```

```python
X
```

```python
#Since we are dealing with distnce metrics we should scale the data
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=2)
```

```python
cluster_labels = model.fit_predict(scaled_X)
```

```python
X['Cluster Labels'] = cluster_labels
```

```python
X
```

```python
X.corr()['Cluster Labels'][:-1].sort_values()
```

```python
plt.figure(figsize=(12,10),dpi=200)
X.corr()['Cluster Labels'][:-1].sort_values().plot(kind='bar')
```

```python
##Choosing a k value
```

```python
ssd = []

for k in range(2,10):
    model = KMeans(n_clusters=k)
    model.fit(X)
    
    ssd.append(model.inertia_)##SSD point ---> Cluster center
```

```python
ssd
```

```python
plt.plot(range(2,10),ssd,'o--')
```

```python
pd.Series(ssd)
```
